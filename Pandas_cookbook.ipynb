### Isin method

```python
# isin series method is equivalent to the SQL IN
depts = ['Houston Police Department-HPD', 'Houston Fire Department (HFD)']
criteria_dept = employee.DEPARTMENT.isin(depts)
criteria_gender = employee.GENDER == 'Female'
criteria_sal = (employee.BASE_SALARY >= 80000) & (employee.BASE_SALARY <= 120000)
criteria_final = (criteria_dept & \
    criteria_gender & \
    criteria_sal)
select_columns = ['UNIQUE_ID', 'DEPARTMENT', 'GENDER', 'BASE_SALARY']
employee.loc[criteria_final, select_columns].head()
```

```python
# isin es el equivalente a IN
# ~ se utiliza para negar
# Select * from EMPLOYEE 
#    WHERE DEPARTMENT NOT IN (
#         SELECT DEPARTMEN FROM (
#               SELECT DEPARTMENT, COUNT(1) AS CI FROM EMPLOYEE GROUP BY DEPARTMENT ORDER BY CI DESC LIMIT 5))

top_5_depts = employee.DEPARTMENT.value_counts().index[:5]
criteria = ~employee.DEPARTMENT.isin(top_5_depts)
employee[criteria]
```

```python
# The filter method applied after a call to the groupby method is 
#    completely different than the DataFrame filter method
college = pd.read_csv(os.path.join(path, 'college.csv'), index_col='INSTNM')
grouped = college.groupby('STABBR')
print(grouped.ngroups)
# Otro método
college['STABBR'].nunique() # verifying the same number
```

```python
# Use the filter method passed with the check_minority function and a
#    threshold of 50% to find all states that have a minority majority:
def check_minority(df, threshold):
    minority_pct = 1 - df['UGDS_WHITE']
    total_minority = (df['UGDS'] * minority_pct).sum()
    total_ugds = df['UGDS'].sum()
    total_minority_pct = total_minority / total_ugds
    return total_minority_pct > threshold

college_filtered = grouped.filter(check_minority, threshold=.5)
college_filtered.head()
```

```python
# Concatenating multiple DataFrames together
stocks_2016 = pd.read_csv(os.path.join(path,'stocks_2016.csv'),
index_col='Symbol')
stocks_2017 = pd.read_csv(os.path.join(path,'stocks_2017.csv'),
index_col='Symbol')
```

```python
# Por defecto concat concatena los dataframes en vertical
s_list = [stocks_2016, stocks_2017]
pd.concat(s_list)
```

```python
# Podemos especificar el origen de la concatenación
pd.concat(s_list, keys=['2016', '2017'],names=['Year', 'Symbol'])
```

```python
# También es posible concatenar horizontalmente
pd.concat(s_list, keys=['2016', '2017'], axis='columns', names=['Year', None])
```

```python
# Podemos usar concat para hacer join entre dos dataframes:
pd.concat(s_list, join='inner', keys=['2016', '2017'],\
    axis='columns', names=['Year', None])
```
